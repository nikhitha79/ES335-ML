```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load the original and reconstructed images
original_img = Image.open('original_image.jpeg')
reconstructed_img = Image.open('reconstructed_image.jpeg')

# Convert images to numpy arrays
original_img_np = np.array(original_img)
reconstructed_img_np = np.array(reconstructed_img)

# Calculate RMSE
mse = np.mean((original_img_np - reconstructed_img_np) ** 2)
rmse = np.sqrt(mse)

# Calculate PSNR
max_pixel_value = 255  # For images in the range [0, 255]
psnr = 20 * np.log10(max_pixel_value / rmse)

# Visualize the images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(original_img_np)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Reconstructed Image')
plt.imshow(reconstructed_img_np)
plt.axis('off')

plt.tight_layout()
plt.show()

print("RMSE:", rmse)
print("PSNR:", psnr)
```
